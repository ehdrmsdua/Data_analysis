```python
midwest = pd.read_csv('C:/Users/USER/OneDrive/바탕 화면/data_analysis/practice/data_file_pr2/midwest.csv')
midwest.head(10)
```

```python
midwest['percent'] = midwest['popasian']/midwest['poptotal'] *100
```

```python
midwest.head()
```

```python
midwest['percent'].sort_values(ascending = False)
```

```python
area = []
for i in midwest['percent']:
    if i >= 0.9:
        area.append('large')
    elif i >= 0.4:
        area.append('middle')
    else: area.append('small')

midwest['new_one'] = area
```

```python
midwest['new_one'].value_counts()
```
