```python
data = {'사람1':[95,100,90,80],'사람2':[85,85,90,100], '사람3':[80,90,80,90] }
indexname = ['국어', '수학', '영어', '탐구']
df = pd.DataFrame(data ,index = indexname)
df = df.T
df
```

```python
df.drop(['탐구'], axis=1, inplace =True)
```

```python
df['평균'] = round((df['국어'] + df['영어']+ df['수학'])/3 ,2)
```

```python
df
```

```python
grades = []
for i in df['평균']:
  if i >= 95:
    grades.append('A+')
  elif i >= 85:
    grades.append('B+')
  else:
    grades.append('B')

df['등급'] = grades
df
```
